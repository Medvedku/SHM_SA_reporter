```python
import datetime
import subprocess

# --- SETTINGS ---
start_week = (2024, 35)   # YYYY, ISO week
end_week   = (2025, 49)

# Path to your wrapper script:
WRAPPER = "scr/plotting/run_full_report.py"

def iso_week_end_date(year, week):
    """Return the Sunday date for an ISO year/week."""
    # ISO weeks: Monday=1 â†’ Sunday=7
    d = datetime.date.fromisocalendar(year, week, 7)
    return d.isoformat()

# Generate list of (year, week)
iso_weeks = []
y, w = start_week

while True:
    iso_weeks.append((y, w))

    # Stop condition
    if (y, w) == end_week:
        break

    # Increment ISO week
    w += 1
    # Handle rollover
    try:
        datetime.date.fromisocalendar(y, w, 1)
    except ValueError:
        y += 1
        w = 1

# --- Run wrapper for each week ---
for (year, week) in iso_weeks:
    end_date = iso_week_end_date(year, week)
    print(f"\n=== Running report for {year}W{week:02d} â†’ END date {end_date} ===")

    cmd = [
        "python3",
        WRAPPER,
        "--end-date", end_date,
        "--yes"
    ]

    print("Command:", " ".join(cmd))
    subprocess.run(cmd, check=True)

print("\nðŸŽ‰ DONE â€” all weekly reports generated!")
```
